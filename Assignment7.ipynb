```python
#0: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
#1: Read data from GitHub using raw URL
url = "https://raw.githubusercontent.com/thytlee/ITEC4700/main/Mall_Customers.csv"
df = pd.read_csv(url)
df.head()
```

```python
#2: Select only the two required columns
data = df[['Annual Income (k$)', 'Spending Score (1-100)']]

data.head()
```

```python
#3: Visualize the raw data
plt.figure(figsize=(7,5))
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Raw Customer Data")
plt.show()
```

```python
#4: Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

scaled_data[:5]
```

```python
#5: Run K-Means with K=3
k3 = KMeans(n_clusters=3, random_state=42)
labels3 = k3.fit_predict(scaled_data)
```

```python
#6: Visualize clusters (K=3)
plt.figure(figsize=(7,5))
plt.scatter(scaled_data[:,0], scaled_data[:,1], c=labels3, cmap='viridis')
plt.title("K-Means Clustering (K=3)")
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.show()
```

```python
#7: Run K-Means with K=4
k4 = KMeans(n_clusters=4, random_state=42)
labels4 = k4.fit_predict(scaled_data)
```

```python
#8: Visualize clusters (K=4)
plt.figure(figsize=(7,5))
plt.scatter(scaled_data[:,0], scaled_data[:,1], c=labels4, cmap='viridis')
plt.title("K-Means Clustering (K=4)")
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.show()
```

```python
#9: Run K-Means with K=5
k5 = KMeans(n_clusters=5, random_state=42)
labels5 = k5.fit_predict(scaled_data)
```

```python
#10: Visualize clusters (K=5)
plt.figure(figsize=(7,5))
plt.scatter(scaled_data[:,0], scaled_data[:,1], c=labels5, cmap='viridis')
plt.title("K-Means Clustering (K=5)")
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.show()
```

```python
#11: Visualize final chosen K=5 again
plt.figure(figsize=(7,5))
plt.scatter(scaled_data[:,0], scaled_data[:,1], c=labels5, cmap='viridis')
plt.title("FINAL CHOICE  K-Means (K=5)")
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.show()
```
